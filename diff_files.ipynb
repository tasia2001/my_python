```python
def normal(x):
    marks=['.',',','-',':',';','-','[',']','\n',""]
    for i in marks:
        x=(str(x).lower()).replace(i,'')
    return x

with open('diff_file1.txt','r',encoding='utf8') as f1, open('diff_file2.txt','r',encoding='utf8') as f2:
    f11=f1.readlines()
    f22=f2.readlines()
    a=[]
    b=[]
    for i in f11:
        a.append(normal(i))
    set1=set((','.join(a)).split(' '))
    for i in f22:
        b.append(normal(i))
    set2=set((','.join(b)).split(' '))
    
    print("Список уникальных слов:",list(set1|set2))
    print("Список слов, входящих в оба файла:",list(set1&set2))
    print("Список слов первого файла, не входящих во второй:", list(set1-set2))
    print("Список слов из второго файла, не входящих в первый:",list(set2-set1))
    print("Список слов, не входящих в оба файла одновременно:",list(set1^set2))
```
