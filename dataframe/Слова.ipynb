```python
import pandas as pd
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
url = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/data_stat/yandex-stat-q.csv"
df = pd.read_csv(url)
df.head(15)
```

```python
def parse(w):
    return morph.parse(w)[0].normal_form
    
    
def fun(s):
    a=map(parse,s.split(" "))
    return " ".join(list(a))
    
```

```python
df['Слова']=df['Поисковая фраза'].apply(fun)
```

Частота упоминания определённых слов в запросах:

```python
pd.Series(' '.join(df['Слова']).split()).value_counts()[:50]
```
